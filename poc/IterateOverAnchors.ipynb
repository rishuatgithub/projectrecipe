```python
url = "https://www.jamieoliver.com/recipes/"
print(url)
```

```python
response = requests.get(url)

response
```

```python
soup = BeautifulSoup(response.text,"html.parser")
```

```python
web_title = soup.title.text

web_title
```

```python
categories = soup.find_all("div",{"class": "tile-wrapper"})
i=0
for tags in categories:
    anchors = tags.findAll('a',href=True)
    
    for aHref in anchors:
        i += 1
        print(i," ",aHref.text.strip()," ",aHref['href'])
        
```

```python
### building the functions:

def printAllHrefs(url,classParam):
    response = requests.get(url)
    soup = BeautifulSoup(response.text,"html.parser")
    categories = soup.find_all("div",{"class": classParam})
        
    #if(param == 1): 
    #    categories = soup.find_all("div",{"class": "tile-wrapper"})
    #else:
    #    categories = soup.find_all("div",{"class": "recipe-block"})
    
    i=0
    for tags in categories:
        anchors = tags.findAll('a',href=True)
        
        for aHref in anchors:
            i += 1
            print(i," ",aHref.text.strip()," ",aHref['href'])
        
```

```python
printAllHrefs("https://www.jamieoliver.com/recipes/","tile-wrapper")
```

```python
printAllHrefs("https://www.jamieoliver.com/recipes/category/world/","recipe-block")
```

```python
printAllHrefs("https://www.jamieoliver.com/recipes/category/world/indian/","recipe-block")
```
